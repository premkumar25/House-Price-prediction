<h2>Import Libraries</h2>

```
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category='FutureWarning')
%matplotlib inline
```

```
df1=pd.read_csv('test.csv')
df1.head()
```

```
df1.info()
```

<h4>Drop fields which has more number of missing values</h4>

```
df1 = df1.drop(['Alley','FireplaceQu','PoolQC','Fence','MiscFeature'], axis=1)
```

<h3>Get the columns which has missing values and number of mising values and column type</h3>

```
for i in df1.columns:
  if df1[i].isnull().sum():
    print(i,df1[i].isnull().sum(),df1[i].dtype)
```

<h3>Fill the missing values with mean and mode</h3>

```
df1['MSZoning'].fillna(df1['MSZoning'].mode()[0],inplace=True)
df1['LotFrontage'].fillna(df1['LotFrontage'].mean(),inplace=True)
df1['Utilities'].fillna(df1['Utilities'].mode()[0],inplace=True)
df1['Exterior1st'].fillna(df1['Exterior1st'].mode()[0],inplace=True)
df1['Exterior2nd'].fillna(df1['Exterior2nd'].mode()[0],inplace=True)
df1['MasVnrType'].fillna(df1['MasVnrType'].mode()[0],inplace=True)
df1['MasVnrArea'].fillna(df1['MasVnrArea'].mean(),inplace=True)
df1['MasVnrArea'].fillna(df1['MasVnrArea'].mode()[0],inplace=True)
df1['BsmtQual'].fillna(df1['BsmtQual'].mode()[0],inplace=True)
df1['BsmtCond'].fillna(df1['BsmtCond'].mode()[0],inplace=True)
df1['BsmtExposure'].fillna(df1['BsmtExposure'].mode()[0],inplace=True)
df1['BsmtFinType1'].fillna(df1['BsmtFinType1'].mode()[0],inplace=True)
df1['BsmtFinSF1'].fillna(df1['BsmtFinSF1'].mean(),inplace=True)
df1['BsmtFinType2'].fillna(df1['BsmtFinType2'].mode()[0],inplace=True)
df1['BsmtFinSF2'].fillna(df1['BsmtFinSF2'].mean(),inplace=True)
df1['BsmtUnfSF'].fillna(df1['BsmtUnfSF'].mean(),inplace=True)
df1['TotalBsmtSF'].fillna(df1['TotalBsmtSF'].mean(),inplace=True)
df1['BsmtFullBath'].fillna(df1['BsmtFullBath'].mean(),inplace=True)
df1['BsmtHalfBath'].fillna(df1['BsmtHalfBath'].mean(),inplace=True)
df1['KitchenQual'].fillna(df1['KitchenQual'].mode()[0],inplace=True)
df1['Functional'].fillna(df1['Functional'].mode()[0],inplace=True)
df1['GarageType'].fillna(df1['GarageType'].mode()[0],inplace=True)
df1['GarageYrBlt'].fillna(df1['GarageYrBlt'].mean(),inplace=True)
df1['GarageFinish'].fillna(df1['GarageFinish'].mode()[0],inplace=True)
df1['GarageCars'].fillna(df1['GarageCars'].mean(),inplace=True)
df1['GarageArea'].fillna(df1['GarageArea'].mean(),inplace=True)
df1['GarageQual'].fillna(df1['GarageQual'].mode()[0],inplace=True)
df1['GarageCond'].fillna(df1['GarageCond'].mode()[0],inplace=True)
df1['SaleType'].fillna(df1['SaleType'].mode()[0],inplace=True)
```

<h3>Plot a Heatmap to find if there are any missing values in the data frame</h3>

```
sns.heatmap(df1.isnull(),yticklabels=False,cbar=False)
```
